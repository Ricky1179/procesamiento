```python
import pandas as pd
```

```python
#Agrupar por sexo, no cuenta los nulos
df.groupby(['sexo'])['clave_alu'].agg('count')
```

```python
#Agrupar por sexo, cuenta los nulos
df.groupby(['sexo'], dropna=False)['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad
df.groupby(['ciudad'])['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad, cuenta los nulos
df.groupby(['ciudad'], dropna=False)['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
df.groupby(['ciudad', 'sexo'], dropna=False)['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
df.groupby(['ciudad', 'sexo'], dropna=False)['calificacion']\
    .agg(['count','sum', 'mean','min', 'max', 'std'])
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
df.groupby(['ciudad', 'sexo'], dropna=False)['calificacion']\
    .agg(['count','sum', 'mean','min', 'max', 'std']).reset_index
df
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
df.groupby('sexo', dropna=False)['calificacion'].mean()
```

```python
#Agrupar por sexo, obtener promedio, filtrado por la cd de QUERETARO
df[df['ciudad']=='QUERETARO'].groupby('sexo', dropna=False)['calificacion'].mean()
```

```python
#Agrupar por sexo, obtener promedio, filtrado por la cd de QUERETARO
df[df['ciudad']=='QUERETARO']\
    .groupby('sexo', dropna=False)['calificacion']\
    .agg(['count','mean'])
```

```python
#Crear una nueva columna
df['calificacion_cadena'] = df['calificacion'].astype(str)
df.info()
```

```python
df.ciudad.\
    str.split(' ')
```

```python
df[['entero','decimal']] = df.calificacion_cadena.\
    str.split('.', expand=True)
df.drop('calificacion_cadena', axis=1, inplace=True)
df
```

```python
df[['usuario','dominio']] = df.email.str.split('@', expand=True)
df
```

```python
# Crear columnas a partir de slice de cadena
df['anio'] = df.curp.str[4:6]
df['mes'] = df.curp.str[6:8]
df['dia'] = df.curp.str[8:10]
df['estado'] = df.curp.str[11:13]
df
```

```python
# El GPA (Grade Point Average) es un término utilizado para asignar un valor numérico 
# a las calificaciones acumuladas 
df['gpa'] = (((df['calificacion']*100)/df['calificacion'].max())/20) -1
df
```

```python
#Generar valor escalar
gpa_rango = [1, 1.7, 2.4, 3.3, 4.1]
gpa_etiqueta = ['D', 'C', 'B', 'A']
df['gpa_escala'] = pd.cut(df['gpa'], bins=gpa_rango, labels=gpa_etiqueta)
print(df['gpa_escala'].unique())
df.info()
```

```python
#Y años, M meses, W semanas, Q trimestres, sms quincena, bms mes
sti = pd.date_range('2023-01-01', periods=df.clave_alu.count(), freq='2w')
print(df.clave_alu.count())
print(sti)
df['fecha'] = sti
```
