```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv")
df
```

```python
# Calcular la media de la edad
mean_age = df['age'].mean()
print(mean_age)
# Llenar los valores nulos con la media de la edad
df['age'].fillna(mean_age, inplace=True)
df
```

```python
df['embarked'].fillna('S', inplace=True)
df
```

```python
# Cambiar los valores alfanuméricos de sex a números
sex_mapping = {'female': 1, 'male': 0}
df['sex'] = df['sex'].replace(sex_mapping)
df
```

```python
# Cambiar los valores alfanuméricos de embarked a números
embarked_mapping = {'S': 0, 'C': 1, 'Q': 2}
df['embarked'] = df['embarked'].replace(embarked_mapping)
df
```

```python
# Crear la nueva variable "child"
df['child'] = df['age'].apply(lambda x: 1 if x < 16 else 0)
df
```

```python
# Eliminar las columnas "ticket", "cabin" y "name"
df.drop(['ticket', 'cabin', 'name'], axis=1, inplace=True)
df
```

```python
# Validar registros duplicados
duplicates = df.duplicated()
# Eliminar registros duplicados
df.drop_duplicates(inplace=True)
df
```

```python
# Listar los valores únicos del campo "sex"
valores_unicos = df['sex'].unique()
print(valores_unicos)
df
```

```python
# Quitar los espacios vacíos de los campos no numéricos
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df
```

```python
# Guardar los datos limpios en una dirección específica
ruta_archivo = "datos/df/iniciales_titanic_clean_AMJR.json"
df.to_json(ruta_archivo, orient='records')
```
